# How to Access a Model from Hugging Face
---

## Method 1: Using the Ready-made `pipeline`

- Visit the Hugging Face piplines:  
   ðŸ”— [https://huggingface.co/docs/transformers/main_classes/pipelines](https://huggingface.co/docs/transformers/main_classes/pipelines)
   
If you want a faster and simpler way to load a model and use it immediately, you can rely on Hugging Face's `pipeline` utility.

Here's how:

```python
from transformers import pipeline

# Create a sentiment analysis pipeline using a pretrained model
classifier = pipeline("sentiment-analysis")

# Use the pipeline to make a prediction
result = classifier("This product is amazing!")
print(result)
```

### ðŸ“Œ Key Points about `pipeline`:

- It automatically loads the correct model and tokenizer based on the task.

- You can specify the model manually (as shown) or let Hugging Face choose a default one.

- It simplifies the entire process â€” no need to separately load models and tokenizers if you use pipelines.

- Pipelines are great for quick demos, prototypes, and experiments

````python
from transformers import pipeline

# 1. Initialize the Text Generation Pipeline
# Using Qwen2.5-1.5B (Excellent for English/Arabic technical content)
generator = pipeline(
    "text-generation",
    model="Qwen/Qwen2.5-1.5B-Instruct",
    device_map="auto"
)

# 2. Initialize the Translation Pipeline
# Using a Helsinki-NLP model for high-speed local translation
translator = pipeline(
    "translation_en_to_ar",
    model="Helsinki-NLP/opus-mt-en-ar"
)

def generate_multilingual_content(topic):
    # Step A: Generate technical content in English
    prompt = f"Write a professional summary about {topic} in telecommunications."
    gen_output = generator(
        prompt,
        max_new_tokens=150,
        do_sample=True,
        temperature=0.7
    )
    english_text = gen_output[0]['generated_text']

    # Step B: Translate the output to Arabic
    translation_output = translator(english_text)
    arabic_text = translation_output[0]['translation_text']

    return english_text, arabic_text

# Example execution
topic_input = "Self-Optimizing Networks (SON) in 5G RAN"
en_content, ar_content = generate_multilingual_content(topic_input)

print(f"--- [EN] ---\n{en_content}\n")
print(f"--- [AR] ---\n{ar_content}")
````

````python
from transformers import pipeline

# 1. Initialize the Summarization Pipeline
# 'facebook/bart-large-cnn' is the gold standard for concise summaries
summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn",
    device_map="auto"
)

def summarize_telecom_report(text):
    # max_length: ensures the summary isn't too long
    # min_length: ensures it's descriptive enough
    summary = summarizer(
        text,
        max_length=130,
        min_length=30,
        do_sample=False
    )
    return summary[0]['summary_text']

# Example: Summarizing a RAN Optimization Insight
technical_text = """
The transition of VoLTE traffic to the L2600 band has resulted in a 15% reduction in congestion
on the lower bands. By utilizing a Self-Optimizing Network (SON) framework, the system can
dynamically adjust Cell Individual Offsets (CIO) and transmission power. This proactive
approach, powered by Deep Reinforcement Learning, ensures that load balancing occurs
before peak traffic thresholds are crossed, maintaining high KPI standards.
"""

print(f"Summary:\n{summarize_telecom_report(technical_text)}")
````

````python
from transformers import pipeline

# 1. Initialize the EN-to-FR Translation Pipeline
# This model is part of the OPUS-MT project, ideal for technical and formal text.
fr_translator = pipeline(
    "translation_en_to_fr",
    model="Helsinki-NLP/opus-mt-en-fr",
    device_map="auto"
)

def translate_to_french(summary_text):
    # Perform the translation
    translation = fr_translator(summary_text)
    return translation[0]['translation_text']

# Input: The summary generated in the previous step
summary_to_translate = (
    "The implementation of a Self-Optimizing Network (SON) framework using Deep "
    "Reinforcement Learning significantly reduces 5G network congestion and improves "
    "load balancing efficiency."
)

french_summary = translate_to_french(summary_to_translate)

print(f"--- [FR] French Translation ---\n{french_summary}")
````

````python
from transformers import pipeline

# 1. Initialize the Question Answering Pipeline
# Using 'deepset/roberta-base-squad2' for its ability to handle "no-answer" scenarios
qa_reader = pipeline(
    "question-answering",
    model="deepset/roberta-base-squad2",
    device_map="auto"
)

def ask_assistant(question, context):
    # The pipeline searches the 'context' for the 'question'
    result = qa_reader(question=question, context=context)

    return {
        "answer": result['answer'],
        "confidence": f"{result['score']:.2%}",
        "location": (result['start'], result['end'])
    }

# Example: Asking about the generated SON report
context_data = """
The Self-Optimizing Network (SON) framework utilizes Deep Reinforcement Learning
to manage load balancing. It achieves a 15% reduction in congestion by dynamically
adjusting Cell Individual Offsets (CIO) and transmission power in the 5G RAN.
"""

user_query = "What is the percentage of congestion reduction?"
answer_details = ask_assistant(user_query, context_data)

print(f"Q: {user_query}")
print(f"A: {answer_details['answer']} (Confidence: {answer_details['confidence']})")
````
